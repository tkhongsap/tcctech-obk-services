```python
client = AzureOpenAI(
        api_version=api_version,
        azure_endpoint=endpoint,
        api_key=subscription_key,
    )

def generate_embeddings(text, model):
    return client.embeddings.create(input = [text], model=model).data[0].embedding

hachiken_true_embedded = generate_embeddings('hachiken ramen', model_name)
starbucks_true_embedded = generate_embeddings('starbucks', model_name)
pizza_true_embedded = generate_embeddings('พิซซ่าคอมปะนี', model_name)
iberry_true_embedded = generate_embeddings('บริษัท iberry จำกัด มหาชน', model_name)

```

```python
hachiken_embedded = []
for case in hachiken_cases:
    hachiken_embedded.append(generate_embeddings(case, model_name))
```

```python
pizza_embedded = []
for case in pizza_cases:
    pizza_embedded.append(generate_embeddings(case, model_name))
```

```python
iberry_embedded = []
for case in iberry_cases:
    iberry_embedded.append(generate_embeddings(case, model_name))
```

```python
starbucks_embedded = []
for case in starbucks_cases:
    starbucks_embedded.append(generate_embeddings(case, model_name))
```

```python
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
model_scores = []
for i in range(len(hachiken_embedded)):
    start_time = time.time()
    score = cosine_similarity(hachiken_true_embedded, hachiken_embedded[i])
    end_time = time.time()
    elapsed = end_time - start_time
    model_scores.append({'test_case': hachiken_cases[i], 'score': float(score) * 100, 'time': elapsed})
```

```python
for i in range(len(starbucks_embedded)):
    start_time = time.time()
    score = cosine_similarity(starbucks_true_embedded, starbucks_embedded[i])
    end_time = time.time()
    elapsed = end_time - start_time
    model_scores.append({'test_case': starbucks_cases[i], 'score': float(score) * 100, 'time': elapsed})
```

```python
for i in range(len(pizza_embedded)):
    start_time = time.time()
    score = cosine_similarity(pizza_true_embedded, pizza_embedded[i])
    end_time = time.time()
    elapsed = end_time - start_time
    model_scores.append({'test_case': pizza_cases[i], 'score': float(score) * 100, 'time': elapsed})
```

```python
for i in range(len(iberry_embedded)):
    start_time = time.time()
    score = cosine_similarity(iberry_true_embedded, iberry_embedded[i])
    end_time = time.time()
    elapsed = end_time - start_time
    model_scores.append({'test_case': iberry_cases[i], 'score': float(score) * 100, 'time': elapsed})
```
